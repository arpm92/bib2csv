```python
import pandas as pd
```

```python
file = open('../00_Input/Mendeley/bib/Library.bib',"r",encoding='utf-8') 
```

```python
txt = file.read()
```

```python
publications = txt.split("@")
```

```python
dicts = [0]*(len(publications))
for i in range (len(publications)):
    dic_pub = {
    "Title": publications[i].partition("\ntitle = {{")[2].partition("}},\n")[0],
    "Author": publications[i].partition("\nauthor = {")[2].partition("},\n")[0],
    "Abstract": publications[i].partition("\nabstract = {")[2].partition("},\n")[0],
    "Keywords": publications[i].partition("\nkeywords = {")[2].partition("},\n")[0],
    "Year": publications[i].partition("\nyear = {")[2].partition("}\n}\n")[0],
    "Journal": publications[i].partition("\njournal = {")[2].partition("},\n")[0],
    "ISSN": publications[i].partition("\nissn = {")[2].partition("},\n")[0],
    "doi": publications[i].partition("\ndoi = {")[2].partition("},\n")[0]
    }
    dicts[i] = dic_pub
```

```python
df = pd.DataFrame(dicts)
```
